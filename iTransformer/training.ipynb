```python
def get_configs(model_name):
    Args = namedtuple('Args', [
        'model_id', 'model', 'data', 'features', 'seq_len', 'label_len', 'pred_len',
        'd_model', 'n_heads', 'e_layers', 'd_layers', 'd_ff', 'factor', 'embed',
        'distil', 'des', 'class_strategy', 'last_zero'
        ])
    # Split the string by underscores
    parts = model_name.split('_')

    # Assign the values to respective variables
    args = Args(
        model_id = parts[0],
        model = parts[1],
        data = parts[2],
        features = parts[3],
        seq_len = int(parts[4][2:]), # 'ft' prefix
        label_len = int(parts[5][2:]), # 'sl' prefix
        pred_len = int(parts[6][2:]),  # 'll' prefix
        d_model = int(parts[7][2:]),  # 'pl' prefix
        n_heads =int(parts[8][2:]),  # 'dm' prefix
        e_layers = int(parts[9][2:]),  # 'nh' prefix
        d_layers = int(parts[10][2:]),  # 'el' prefix
        d_ff = int(parts[11][2:]),  # 'dl' prefix
        factor = int(parts[12][2:]),  # 'df' prefix
        embed = parts[13][2:],  # 'fc' prefix
        distil = parts[14][2:],  # 'eb' prefix
        des = parts[15][2:],  # 'dt' prefix
        class_strategy = parts[16],
        last_zero = int(parts[17])  # This is always '0'
    )
    return args
```
